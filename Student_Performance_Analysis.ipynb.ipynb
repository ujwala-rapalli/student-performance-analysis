```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
# Display the first 5 rows
df.head()
```

```python
# Dataset structure and summary
print("Number of rows and columns:", df.shape)
print("\nData types and non-null counts:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
```

```python
# Mean, median, and standard deviation for G1, G2, and G3 (Grades)
scores_stats = df[['G1', 'G2', 'G3']].agg(['mean', 'median', 'std'])
print(scores_stats)
```

```python
# Minimum and maximum scores for each of G1, G2, G3
scores_min_max = df[['G1', 'G2', 'G3']].agg(['min', 'max'])
print(scores_min_max)
```

```python
#Data visualization
# Set up for visual consistency
plt.figure(figsize=(15, 5))
scores = ['G1', 'G2', 'G3']

for i, score in enumerate(scores, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[score], bins=10, kde=True)
    plt.title(f'Distribution of {score}')
    plt.xlabel(score)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Average scores by gender (sex column)
avg_scores_by_gender = df.groupby('sex')[['G1', 'G2', 'G3']].mean()
avg_scores_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='G1', y='G2', data=df)
plt.title("Relationship between G1 and G2 Scores")
plt.xlabel("G1 Score")
plt.ylabel("G2 Score")
plt.show()
```

```python
# Average scores by mother's education (Medu) and father's education (Fedu)
avg_scores_by_medu = df.groupby('Medu')[['G1', 'G2', 'G3']].mean()
avg_scores_by_fedu = df.groupby('Fedu')[['G1', 'G2', 'G3']].mean()

# Plotting
avg_scores_by_medu.plot(kind='bar', figsize=(12, 6))
plt.title("Average Scores by Mother's Education Level")
plt.ylabel("Average Score")
plt.xlabel("Mother's Education Level")
plt.xticks(rotation=0)
plt.show()

avg_scores_by_fedu.plot(kind='bar', figsize=(12, 6))
plt.title("Average Scores by Father's Education Level")
plt.ylabel("Average Score")
plt.xlabel("Father's Education Level")
plt.xticks(rotation=0)
plt.show()
```

```python
# Average scores based on test preparation course (Pstatus)
avg_scores_by_test_preparation = df.groupby('Pstatus')[['G1', 'G2', 'G3']].mean()
avg_scores_by_test_preparation.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Test Preparation Status")
plt.ylabel("Average Score")
plt.xlabel("Test Preparation Status")
plt.xticks(rotation=0)
plt.show()
```
